# Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй адаптированные RFM-кластеры студентов, чтобы качественно оценить свою аудиторию.  В адаптированной кластеризации можешь выбрать следующие метрики: R - среднее время сдачи одного экзамена, F - завершаемость курсов, M - среднее количество баллов, получаемое за экзамен. Подробно опиши, как ты создавал кластеры.  Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. 

```python
#Построим такие кластеры студентов 
rfm2 = exsclist.query('assessment_type == "Exam"')\
    .groupby('id_student',as_index = False)\
    .agg({'date_submitted':'mean', 'score': 'mean'})
#Основой таблицы станет уже неоднократно пригодившаяся нам таблица из задания №1, и сразу посчитаем среднее количество дней
# потраченное на успешную сдачу экзаменов студентами и среднее набранное количество баллов
```

```python
a = exsclist.query('score >= 40 and assessment_type == "Exam"')\
    .groupby('id_student', as_index = False)\
    .agg({'score':'count'})  
```

```python
b = exsclist.query('assessment_type == "Exam"')\
    .groupby('id_student', as_index = False)\
    .agg({'score':'count'})
#Для подсчета завершаемости сделаем две отдельных небольших таблицы со студентами и их количесвтом СДАННЫХ экзаменов
# и студентами и количеством СДАВАЕМЫМ ими экзаменов
```

```python
ab = a.merge(b, on = 'id_student', how = 'right')
```

```python
rfm2['completion'] = ab.score_x / ab.score_y #объединим и посчитаем завершаемость
```

```python
rfm2 = rfm2.fillna(0)
#Ввиду того, что в таблице a нашлось место не всем студентам из таблицы b (студенты так и не сдавшие свои экзамены),
#заполняем NaN`ы "0", как индикатор нулевой завершемости курсов студентом
rfm2
```

```python
rfm2.completion.mean() - 2.576 * np.std(rfm2.completion), rfm2.completion.mean() + 2.576 * np.std(rfm2.completion)
#границы средней завершаемости сдачи экзаменов
```

```python
rfm2.score.mean() - 2.576 * np.std(rfm2.score), rfm2.score.mean() + 2.576 * np.std(rfm2.score)
#границы средних баллов сдачи экзаменов
```

```python
rfm2.date_submitted.mean() - 2.576 * np.std(rfm2.date_submitted), rfm2.date_submitted.mean() + 2.576 * np.std(rfm2.date_submitted)
#границы средних дат сдачи экзаменов
```

```python
quintiles = rfm2[['date_submitted', 'completion', 'score']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
#Расчитаем квантили для ориентации в основных показателях и количестве их набравшем/преодолевшем
```

```python
#Создадим функции для присвоения RFM меток студентам
def date_score(x):
    if x <= quintiles['date_submitted'][.2]:
        return 5
    elif x <= quintiles['date_submitted'][.4]:
        return 4
    elif x <= quintiles['date_submitted'][.6]:
        return 3
    elif x <= quintiles['date_submitted'][.8]:
        return 2
    else:
        return 1

def comp_score(x):
    if x < 1:
        return 0
    else:
        return 1

def sc_score(x):
    if x < quintiles['score'][.2]:
        return 1
    elif x < quintiles['score'][.4]:
        return 2
    elif x < quintiles['score'][.6]:
        return 3
    elif x < quintiles['score'][.8]:
        return 4
    else:
        return 5    
    
rfm2['R'] = rfm2['date_submitted'].apply(lambda x: date_score(x))
rfm2['F'] = rfm2['completion'].apply(lambda x: comp_score(x))
rfm2['M'] = rfm2['score'].apply(lambda x: sc_score(x))
rfm2
```

```python
rfm2['RFM Score'] = rfm2['R'].map(str) + rfm2['F'].map(str) + rfm2['M'].map(str)
rfm2.head()
```

```python
segt_map = {
    r'[1-5][0][1]': 'F ',
    r'[1-5][0-1][1-2]': 'D ',
    r'[1-5][0-1][2-3]': 'C ',
    r'[1-5][0-1][3-4]': 'B ',
    r'[1-5][1][4-5]': 'A '
}

#Разделим студентов на основные категории:
#А (отлично), В (хорошо), С (удовлетворительно), D (слабо) и F (неудовлетворительно). `

rfm2['Segment'] = rfm2['R'].map(str) + rfm2['F'].map(str) + rfm2['M'].map(str)
rfm2['Segment'] = rfm2['Segment'].replace(segt_map, regex=True)
rfm2.head()
```
